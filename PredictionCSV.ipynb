```python
# Skrpit für die Predictions.csv 

import os
import pandas as pd

# Pfad zu dem Ordner, der die Bilder und Textdateien enthält
image_folder = r"C:\Users\asus\Desktop\SENCE\2. Semester\Projektarbeit 2\01. Abschluss\Abschluss_predictions"

# Definieren Sie die Kategorien
categories = ['car', 'entrances in houses', 'marked parking space', 
              'marked parking space road', 'moving cars', 'parked roadside', 'unparked roadside']

# Erstellen einer leeren Liste, um die Daten für die Tabelle zu speichern
data = []

# Durchsuchen des Ordners und Auflisten aller Bild-IDs und zugehörigen Kategorien
for file_name in os.listdir(image_folder):
    if file_name.endswith('.png'):
        # Bild-ID aus dem Dateinamen extrahieren
        image_id = os.path.splitext(file_name)[0]
        
        # Pfad zur zugehörigen .txt Datei
        txt_file_path = os.path.join(image_folder, image_id + '.txt')
        
        # Initialisiere die Zählungen der tatsächlichen Kategorien mit Nullen
        actual_categories_count = [0] * 7
        
        if os.path.exists(txt_file_path):
            with open(txt_file_path, 'r') as file:
                lines = file.readlines()
                for line in lines:
                    line = line.strip()
                    if line in categories:
                        actual_categories_count[categories.index(line)] += 1

        # Fügen Sie eine neue Zeile mit den Bildinformationen hinzu
        data.append({
            'Bild-ID': image_id,
            'Noteb. 1 (car)': actual_categories_count[0],
            'Noteb. 2 (entrances in houses)': actual_categories_count[1],
            'Noteb. 3 (marked parking space)': actual_categories_count[2],
            'Noteb. 4 (marked parking space road)': actual_categories_count[3],
            'Noteb. 5 (moving cars)': actual_categories_count[4],
            'Noteb. 6 (parked roadside)': actual_categories_count[5],
            'Noteb. 7 (unparked roadside)': actual_categories_count[6],
            'Checked 1 (car)': '',  # Diese Felder bleiben leer, bis das Modell ausgeführt wird
            'Checked 2 (entrances in houses)': '',
            'Checked 3 (marked parking space)': '',
            'Checked 4 (marked parking space road)': '',
            'Checked 5 (moving cars)': '',
            'Checked 6 (parked roadside)': '',
            'Checked 7 (unparked roadside)': '',
            'Richtig/Falsch': '',
            'Fehlerbeschreibung': ''
        })

# Erstellen eines DataFrame aus der Liste
df = pd.DataFrame(data)

# Ausgabe des DataFrames in Python (direkt in Jupyter Notebook oder Konsole)
print(df)

# Optional: Speichern der Tabelle als CSV-Datei mit einem geeigneten Trennzeichen (Komma)
output_csv = 'AbschlussPrediction.csv'
df.to_csv(output_csv, index=False, sep=';')  # Verwenden Sie das Komma als Trennzeichen

print(f"Tabelle wurde erstellt und in {output_csv} gespeichert.")

```
